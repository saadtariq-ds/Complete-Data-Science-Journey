## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, KMeans
```

## Import Dataset

### Blobs Dataset

```python
blobs = pd.read_csv('cluster_blobs.csv')
blobs.head()
```

```python
sns.scatterplot(data=blobs, x='X1', y='X2')
plt.show()
```

### Moon Dataset

```python
moons = pd.read_csv('cluster_moons.csv')
moons.head()
```

```python
sns.scatterplot(data=moons, x='X1', y='X2')
plt.show()
```

### Circles Dataset

```python
circles = pd.read_csv('cluster_circles.csv')
circles.head()
```

```python
sns.scatterplot(data=circles, x='X1', y='X2')
plt.show()
```

## Comparison

```python
def display_categories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(
        data=data,
        x='X1',
        y='X2',
        hue=labels,
        palette='viridis'
    )
```

### KMeans Clustering

```python
kmeans_model = KMeans(n_clusters=3)
```

```python
display_categories(model=kmeans_model, data=blobs)
```

```python
kmeans_model = KMeans(n_clusters=2)
```

```python
display_categories(model=kmeans_model, data=moons)
```

```python
kmeans_model = KMeans(n_clusters=2)
```

```python
display_categories(model=kmeans_model, data=circles)
```

### DBSCAN Clustering

```python
dbscan_model = DBSCAN()
```

```python
display_categories(model=dbscan_model, data=blobs)
```

```python
dbscan_model = DBSCAN(eps=0.15)
```

```python
display_categories(model=dbscan_model, data=moons)
```

```python
dbscan_model = DBSCAN(eps=0.15)
```

```python
display_categories(model=dbscan_model, data=circles)
```
