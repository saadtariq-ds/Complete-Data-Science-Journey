## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import cross_validate
```

## Importing Dataset

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Data Preparaing

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Preprocessing

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

### Model 1

```python
model_one = Ridge(alpha=100)
```

```python
scores = cross_validate(model_one,
                        scaled_X_train,
                        y_train,
                        scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'],
                        cv=10)
```

```python
scores_dataframe = pd.DataFrame(scores)
```

```python
scores_dataframe
```

```python
scores_dataframe.mean()
```

### Model 2

```python
model_two = Ridge(alpha=1)
```

```python
scores = cross_validate(model_two,
                        scaled_X_train,
                        y_train,
                        scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'],
                        cv=10)
```

```python
scores_dataframe = pd.DataFrame(scores)
```

```python
scores_dataframe
```

```python
scores_dataframe.mean()
```

### Final Evaluation

```python
model_two.fit(scaled_X_train, y_train)
```

```python
final_predictions = model_two.predict(scaled_X_test)
```

```python
print(f"MAE : {mean_absolute_error(y_test, final_predictions)}")
print(f"MSE : {mean_squared_error(y_test, final_predictions)}")
print(f"RMSE : {np.sqrt(mean_squared_error(y_test, final_predictions))}")
```
