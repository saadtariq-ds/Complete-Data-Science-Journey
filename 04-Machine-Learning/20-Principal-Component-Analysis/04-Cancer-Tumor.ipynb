## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')
```

## Import Dataset

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
df.head()
```

```python
df.columns
```

## Standardization

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean()
```

```python
scaled_X.mean(axis=0)
```

## Principal Component Analysis

```python
model = PCA(n_components=2)
model.fit_transform(scaled_X)
```

```python
principal_components = model.fit_transform(scaled_X)
```

### Model Components

```python
model.components_
```

```python
df_components = pd.DataFrame(
    data=model.components_,
    index=['PC1', 'PC2'],
    columns=df.columns
)
df_components
```

```python
plt.figure(figsize=(20, 3), dpi=150)
sns.heatmap(df_components, annot=True)
plt.show()
```

### Explained Variance Ratio

```python
model.explained_variance_ratio_
```

```python
np.sum(model.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1, 30):
    model = PCA(n_components=n)
    model.fit(scaled_X)
    explained_variance.append(np.sum(model.explained_variance_ratio_))
```

```python
plt.plot(range(1, 30), explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained")
```
