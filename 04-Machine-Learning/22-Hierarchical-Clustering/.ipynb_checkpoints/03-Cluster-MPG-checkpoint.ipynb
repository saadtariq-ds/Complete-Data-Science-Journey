## Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import scipy.cluster.hierarchy as sc
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

## Import Dataset

```python
df = pd.read_csv('cluster_mpg.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## Data Preprocessing

### Dummy Variables

```python
df['origin'].value_counts()
```

```python
df_dummies = pd.get_dummies(df.drop('name', axis=1))
df_dummies.head()
```

### Standardization

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_dummies)
```

```python
scaled_df = pd.DataFrame(
    data=scaled_data,
    columns=df_dummies.columns
)

scaled_df.head()
```

## Exploratory Data Analysis

```python
plt.figure(figsize=(15, 8), dpi=200)
sns.heatmap(scaled_df)
plt.show()
```

```python
plt.figure(figsize=(15, 8), dpi=200)
sns.clustermap(scaled_df, row_cluster=False)
plt.show()
```

```python
plt.figure(figsize=(15, 8), dpi=200)
sns.clustermap(scaled_df, col_cluster=False)
plt.show()
```

```python
plt.figure(figsize=(15, 8), dpi=200)
sns.heatmap(scaled_df.corr(), annot=True)
plt.show()
```

## Hierarchical Clustering

```python
model = AgglomerativeClustering(
    n_clusters=4
)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
# cluster_labels
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels, palette='viridis')
plt.show()
```

### Max Distance Possible

### New Model

```python
model = AgglomerativeClustering(
    n_clusters=None,
    distance_threshold=0
)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
# cluster_labels
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels, palette='viridis')
plt.show()
```

### Dendogram

```python
linkage_matrix = hierarchy.linkage(
    model.children_
)
linkage_matrix
```

```python
plt.figure(figsize=(20, 10), dpi=200)
dendro = dendrogram(linkage_matrix)
plt.show()
```

```python
plt.figure(figsize=(20, 10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode='lastp', p=10)
plt.show()
```

```python
plt.figure(figsize=(20, 10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode='level', p=3)
plt.show()
```
