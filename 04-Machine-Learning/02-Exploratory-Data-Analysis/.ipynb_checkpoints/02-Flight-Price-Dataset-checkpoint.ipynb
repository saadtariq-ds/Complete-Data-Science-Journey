## FEATURES
The various features of the cleaned dataset are explained below:
1) Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
10) Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
11) Price: Target variable stores information of the ticket price.

## Feature Engineering and EDA

```python
df.head()
```

### Feature Engineering on Date of Journey

```python
df['date'] = (df['Date_of_Journey'].str.split('/').str[0]).astype(int)
df['month'] = (df['Date_of_Journey'].str.split('/').str[1]).astype(int)
df['year'] = (df['Date_of_Journey'].str.split('/').str[2]).astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Date_of_Journey', axis=1)
df.head()
```

### Feature Engineering on Arrival Time

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['arrival_hour'] = (df['Arrival_Time'].str.split(':').str[0]).astype(int)
df['arrival_minute'] = (df['Arrival_Time'].str.split(':').str[1]).astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Arrival_Time', axis=1)
df.head()
```

### Feature Engineering on Departure Time

```python
df['departure_hour'] = (df['Dep_Time'].str.split(':').str[0]).astype(int)
df['departure_minute'] = (df['Dep_Time'].str.split(':').str[1]).astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Dep_Time', axis=1)
df.head()
```

### Feature Engineering on Duration

```python
df['duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['duration_minute'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]

df.head()
```

```python
df['duration_minute'] = (df['duration_minute'].fillna(0)).astype(int)
df.head()
```

```python
df.info()
```

```python
df = df.drop('Duration', axis=1)
df.head()
```

### Feature Engineering on Total Stops

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop':0, '1 stop':1, '2 stops':2,
    '3 stops':3, '4 stops':4, np.nan:1
})

df.head()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.info()
```

```python
df.head()
```

### Feature Engineering on Route

Dropping this column as we already have Source and Destination Columns

```python
df = df.drop('Route', axis=1)
df.head()
```

### Feature Engineering on Airline, Source and Destination

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
df['Destination'].unique()
```

```python
one_hot_encoder = OneHotEncoder()
one_hot_encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray()
```

```python
encoder_df = pd.DataFrame(
    one_hot_encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray(), 
    columns=one_hot_encoder.get_feature_names_out()
)

encoder_df.head()
```

```python
df = pd.concat([df, encoder_df], axis=1)
df.head()
```

```python
df.info()
```

```python
df = df.drop(['Airline', 'Source', 'Destination'], axis=1)
df.head()
```
