```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
```

```python
X = 6 * np.random.rand(100, 1) - 3
y = 0.5 * X**2 + 1.5*X + 2 + np.random.randn(100, 1)
```

```python
plt.scatter(X, y, color='g')
plt.xlabel('X Dataset')
plt.ylabel('Y Dataset')
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

## Simple Linear Regression

```python
simple_regression = LinearRegression()
```

```python
simple_regression.fit(X_train, y_train)
```

```python
simple_regression_prediction = simple_regression.predict(X_test)
```

```python
simple_regression_r_score = r2_score(y_true=y_test, y_pred=simple_regression_prediction)

print(f"R-squared Score: {simple_regression_r_score}")
```

```python
plt.plot(X_train, simple_regression.predict(X_train), color='r')
plt.scatter(X_train, y_train)
```

## Polynomial Regression

```python
polynomial_features = PolynomialFeatures(degree=2, include_bias=True)
```

```python
X_train_polynomial = polynomial_features.fit_transform(X_train)
X_test_polynomial = polynomial_features.fit_transform(X_test)
```

```python
polynomial_regression = LinearRegression()
polynomial_regression.fit(X_train_polynomial, y_train)
polynomial_regression_prediction = polynomial_regression.predict(X_test_polynomial)

polynomial_regression_r_score = r2_score(y_true=y_test, y_pred=polynomial_regression_prediction)
print(f"R-squared Score: {polynomial_regression_r_score}")
```

```python
print(f"Coefficient or slope: {polynomial_regression.coef_}")
print(f"Intercept: {polynomial_regression.intercept_}")
```

```python
plt.scatter(X_train, polynomial_regression.predict(X_train_polynomial), color='r')
plt.scatter(X_train, y_train)
```

### Lets say degree = 3

```python
polynomial_features = PolynomialFeatures(degree=3, include_bias=True)
X_train_polynomial = polynomial_features.fit_transform(X_train)
X_test_polynomial = polynomial_features.fit_transform(X_test)

polynomial_regression = LinearRegression()
polynomial_regression.fit(X_train_polynomial, y_train)
polynomial_regression_prediction = polynomial_regression.predict(X_test_polynomial)

polynomial_regression_r_score = r2_score(y_true=y_test, y_pred=polynomial_regression_prediction)
print(f"R-squared Score: {polynomial_regression_r_score}")

print(f"Coefficient or slope: {polynomial_regression.coef_}")
print(f"Intercept: {polynomial_regression.intercept_}")
```

```python
plt.scatter(X_train, polynomial_regression.predict(X_train_polynomial), color='r')
plt.scatter(X_train, y_train)
```

## Prediction of New Data

```python
X_new = np.linspace(-3, 3, 200).reshape(200, 1)
X_new_polynomial = polynomial_features.transform(X_new)
```

```python
y_new = polynomial_regression.predict(X_new_polynomial)
plt.plot(X_new, y_new, "r-", linewidth=2, label=" New Predictions")
plt.plot(X_train, y_train, "b.",label='Training points')
plt.plot(X_test, y_test, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Polynomial Regression Pipeline

```python
def poly_regression(degree):
    X_new = np.linspace(-3, 3, 200).reshape(200, 1)
    
    polynomial_features=PolynomialFeatures(degree=degree,include_bias=True)
    regression=LinearRegression()
    polynomial_regression=Pipeline([
        ("poly_features",polynomial_features),
        ("lin_reg",regression)
    ])
    
    polynomial_regression.fit(X_train,y_train) ## ploynomial and fit of linear reression
    y_pred_new=polynomial_regression.predict(X_new)
    
    plt.plot(X_new, y_pred_new, "r-", linewidth=2, label=" Degree" + str(degree))
    plt.plot(X_train, y_train, "b.",label='Training points', linewidth=3)
    plt.plot(X_test, y_test, "g.",label='Testing points', linewidth=3)
    
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4, 0, 10])
    plt.show()
```

```python
poly_regression(1)
```
