## Data Preparation

```python
df.head()
```

### Dummy Variables

```python
X = pd.get_dummies(df)
X
```

### Standardization

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

## K Means Clustering

```python
model = KMeans(n_clusters=2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

### Joining Cluster Labels with X

```python
X['Cluster'] = cluster_labels
```

### Checking Correlation

```python
plt.figure(figsize=(12,8), dpi=150)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
```

## Elbow Method

```python
ssd = []
for k in range(2, 15):
    k_means = KMeans(n_clusters=k, init='k-means++')
    k_means.fit(scaled_X)
    ssd.append(k_means.inertia_)
```

```python
plt.plot(range(2, 15), ssd)
plt.xticks(range(2, 15))
plt.xlabel("Number of Clusters")
plt.ylabel("Sum of Squared Distances")
plt.show()
```

## Knee Locator

```python
knee_locator = KneeLocator(range(2,15), ssd, curve="convex", direction="decreasing")
knee_locator.elbow
```

## Silhouette Score

```python
silhouette_coefficients = []

for k in range(2, 15):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(scaled_X)
    score = silhouette_score(scaled_X, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2, 15), silhouette_coefficients)
plt.xticks(range(2, 15))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```
