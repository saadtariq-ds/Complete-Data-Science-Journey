## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
from sklearn.cluster import KMeans
```

## Read Image

```python
image = mpimg.imread('palm_trees.jpg')
image
```

```python
image.shape   # (H, W, C)
```

## Displaying Image

```python
plt.imshow(image)
```

## K Means Quantization

```python
image.shape
```

```python
(h, w, c) = image.shape

print(f"Image Height: {h}")
print(f"Image Width: {w}")
print(f"Image Color: {c}")
```

```python
image_as_2d_array = image.reshape(h*w, c)
image_as_2d_array
```

```python
len(image.shape)
```

```python
len(image_as_2d_array.shape)
```

## K Means Model

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(image_as_2d_array)
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
rgb_codes
```

```python
rgb_codes[labels]
```

```python
quantized_image = np.reshape(rgb_codes[labels], (h,w,c))
quantized_image
```

```python
plt.imshow(quantized_image)
```
