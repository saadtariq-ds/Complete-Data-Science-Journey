```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
```

```python
X, y = make_blobs(
    n_samples=1000,
    centers=3,
    n_features=2
)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, palette='viridis')
```

## Standardization

```python
scaler = StandardScaler()
```

```python
X_train, X_test, _, _ = train_test_split(
    X, y, test_size=0.33, random_state=101
)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Modeling

### Elbow Method

```python
wcss = []
for k in range(1, 11):
    k_means = KMeans(n_clusters=k, init='k-means++')
    k_means.fit(X_train)
    wcss.append(k_means.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

### K Means Clustering

```python
model = KMeans(n_clusters=3, init='k-means++')
model.fit_predict(X_train)
```

```python
y_pred = model.predict(X_test)
```

```python
sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=y_pred, palette='viridis')
```

## Validating K Values

### Knee Locator

```python
knee_locator = KneeLocator(range(1,11), wcss, curve="convex", direction="decreasing")
knee_locator.elbow
```

### Silhouetee Scoring

```python
silhouette_coefficients = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X_train)
    score = silhouette_score(X_train, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```
