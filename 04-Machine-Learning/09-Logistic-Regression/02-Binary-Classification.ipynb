```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

## Create Dataset

```python
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

## Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

## Prediction

```python
y_pred = model.predict(X_test)
```

```python
# model.predict_proba(X_test)
```

## Evaluation

```python
accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
print(f"Accuracy: {accuracy}")
```

```python
matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
matrix
```

```python
print(classification_report(y_true=y_test, y_pred=y_pred))
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(matrix, annot=True, fmt='0.1f')
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(matrix/np.sum(matrix), annot=True, 
            fmt='.2%', cmap='Blues')
```

## HyperParameter Tuning and Cross Validation

```python
model = LogisticRegression()
```

```python
penalty = ['l1', 'l2', 'elasticnet']
C = [100, 10, 1.0, 0.1, 0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params = dict(penalty=penalty,C=C, solver=solver)
```

### Grid Search CV

```python
cv = StratifiedKFold(n_splits=5)
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
print(f"Accuracy: {accuracy}")
```

```python
print(classification_report(y_true=y_test, y_pred=y_pred))
```

```python
plt.figure(figsize=(4,3))
matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
sns.heatmap(matrix, annot=True, fmt='0.1f')
```

### Randomized SearchCV

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
C = [100, 10, 1.0, 0.1, 0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
params = dict(penalty=penalty,C=C, solver=solver)

random_cv = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')
```

```python
random_cv.fit(X_train, y_train)
```

```python
random_cv.best_score_
```

```python
random_cv.best_params_
```

```python
y_pred = random_cv.predict(X_test)
```

```python
accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
print(f"Accuracy: {accuracy}")
```

```python
print(classification_report(y_true=y_test, y_pred=y_pred))
```

```python
plt.figure(figsize=(4,3))
matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
sns.heatmap(matrix, annot=True, fmt='0.1f')
```
