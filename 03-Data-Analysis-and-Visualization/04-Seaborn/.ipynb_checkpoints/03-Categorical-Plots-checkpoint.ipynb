# Categorical Plots

Categorical plots displays a statistical metrics per a category.

For example mean value per category or a count of the number of rows per category

There are two main type of plots
1. Countplot(): count number of rows per category
2. barplot(): General form of displaying any chosen metric per category

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

## Countplot

```python
df['division'].value_counts()
```

```python
sns.countplot(data=df, x='division')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='division')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='level of education')
```

### Different Parameters

**Hue Parameter**

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='level of education', hue='division')
```

**Color Mapping**

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='level of education', hue='division', palette="Set3")
```

## Barplot

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary')
```

### Different Parameters

**Estimator**

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean)
```

**Confidence Interval**

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', errorbar="sd", estimator=np.mean, hue='division')

plt.legend(loc=(1.1,0.5))
```

## Distribution within Categories

1. Boxplot
2. Violinplot
3. Swarmplot
4. Boxenplot (Letter-Value Plot)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

### Boxplot

The boxplot displays the distribution of a continuous variable. It does this through the use of quartiles.

Quartiles separate out the data into 4 equal number of data points.
1. 25% of data points are in bottom quartile
2. 50th percentile (Q2) is the median

```python
plt.figure(figsize=(6,4), dpi=100)
sns.boxplot(data=df, y='math score')
```

```python
plt.figure(figsize=(6,4), dpi=100)
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(6,4), dpi=100)
sns.boxplot(data=df, y='reading score', x='test preparation course')
```

#### Different Parameters

**Hue Parameter**

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x='parental level of education', hue='test preparation course')

plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.boxplot(data=df, x='reading score', y='parental level of education', hue='test preparation course')

plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

**Palette Parameter**

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x='parental level of education', hue='test preparation course', palette='Set2')

plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

### Violinplots

The violin plots plays a similar role as the box plot.

It displays the probability density across the data using a KDE.

It can be imagine as a mirrored KDE plot

```python
plt.figure(figsize=(12,3), dpi=200)
sns.violinplot(data=df, y='reading score', x='parental level of education')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education')
```

#### Different Parameters

**Hue and Palette Parameters**

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='reading score', x='parental level of education', hue='test preparation course', palette='Set2')

plt.legend(bbox_to_anchor=(1.0, 0.5))
```

**Split Parameter**

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='reading score', x='parental level of education', hue='test preparation course', palette='Set2',
              split=True)

plt.legend(bbox_to_anchor=(1.0, 0.5))
```

**Inner Parameter**

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='reading score', x='parental level of education', hue='test preparation course', palette='Set2',
              split=True, inner=None)

plt.legend(bbox_to_anchor=(1.0, 0.5))
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='reading score', x='parental level of education', hue='test preparation course', palette='Set2',
              split=True, inner='quartile')

plt.legend(bbox_to_anchor=(1.0, 0.5))
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education',
              split=True, inner='quartile')
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education',
              split=True, inner='stick')
```

**Bandwidth Parameter**

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', bw=0.1)
```

### Swarmplot

The swarmplot is very simple and simply shows all the data points in the distribution.

For very large data sets, it won't show all the points, but will display the general distribution of them

```python
plt.figure(figsize=(4,3), dpi=200)
sns.swarmplot(data=df, x='math score')
```

```python
plt.figure(figsize=(4,3), dpi=200)
sns.swarmplot(data=df, x='math score', size=2)
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course')

plt.legend(bbox_to_anchor=(1.35, 0.5))
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.35, 0.5))
```

### Boxenplot

The boxenplot is a designed as an expansion upon normal box plot

```python
sns.boxenplot(data=df, x='math score', y='test preparation course')
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender')
```
